```python
from ultralytics import YOLO

# Load a pretrained YOLO26n model
model = YOLO("yolo26n.pt")
# detection

results = model("animals.jpg")
results[0].show()
```

```python
results1 = model("dog_cat.jpg")
results1[0].show()
```

#viewing in code

```python
import matplotlib.pyplot as plt

results1 = model("dog_cat.jpg")

img = results1[0].plot()   # returns annotated image (numpy array)

plt.imshow(img)
plt.axis("off")
```
